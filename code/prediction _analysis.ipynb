```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_2_3_8_oof_seed_1 = pd.read_csv('../weights/version_v2_3_11_5_2/oof_seed_1.csv')
train_df = pd.read_csv('../data/train_fold_v6.csv')
```

```python
print('OOF ',df_2_3_8_oof_seed_1.shape)
print('train_df ',train_df.shape)
df_2_3_8_oof_seed_1.head()
```

```python

act_label = [str(i) for i in range(19)]
pred_label = [f'pred_{i}' for i in range(19)]
df_labels_actual = df_2_3_8_oof_seed_1[act_label]
df_labels_pred = df_2_3_8_oof_seed_1[pred_label]
coocc = df_labels_actual.T.dot(df_labels_pred)
```

```python
coocc
```

```python
coocc
```

```python
fig, ax = plt.subplots(figsize=(16, 13))
sns.heatmap(coocc, cmap="Blues", linewidth=0.3, cbar_kws={"shrink": .8})
title = 'How often do individual classes cooccur in the train set?\n(Single class occurrences are on the diagonal)\n'
plt.title(title, loc='center', fontsize=18)
plt.show()
```
